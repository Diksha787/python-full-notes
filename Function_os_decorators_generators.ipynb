**Write a Python function to check whether a string is a pangram or not.**

```python
def pangram(s):
    check=""
    small=s.lower()
    combine=small.replace(" ","")
    for i in combine:
        if i in check:
            return False
        else:
            check+=i
    return True
print(pangram("The quick brown fox jumps over the lazy dog"))
```

**54.Given a sorted array of positive integers. Write a function to rearrange the array elements alternatively i.e first element should be max value, second should be min value, third should be second max, fourth should be second min and so on.**

```python
def rearrange(arr, n): 
        
        a =[]
        
        
        l = 0
        k = n-1
        for i in range(n):
            if (i %2 == 0):
                a.append(arr[k])
                k = k-1
            else :
                a.append(arr[l])
                l = l+1
        for i in range(n):
            arr[i] = a[i]
        return arr

arr = [1,2,3,4,5,6]
N = 6
rearrange(arr, N)
```

```python
def function_decorator(random_func):
  def wrapped_func():
    print("+"*30)
    random_func()
    print("-"*30)
  return wrapped_func
```

```python
@function_decorator
def new_func():
  print("Hey whatsup")
```

```python
function_decorator(new_func)
```
