**global Keyword**

```python
x = "awesome"

def myfunc():
  global x
  x = "fantastic"

myfunc()

print("Python is " + x)
```

```python
def f():
    global s
    s += ' GFG'
    print(s)
    s = "Look for Geeksforgeeks Python Section"
    print(s)
 
# Global Scope
s = "Python is great!"
f()
print(s)
```

```python

a = 1
 
# Uses global because there is no local 'a'
def f():
    print('Inside f() : ', a)
 
# Variable 'a' is redefined as a local
def g():
    a = 2
    print('Inside g() : ', a)
 
# Uses global keyword to modify global 'a'
def h():
    global a
    a = 3
    print('Inside h() : ', a)
 
 
# Global scope
print('global : ', a)
f()
print('global : ', a)
g()
print('global : ', a)
h()
print('global : ', a)
```

```python
def funargs(normal,*asargument):
  print(normal)
  for item in asargument:
    print(item)

text=["python","Java","c","C++","ML","DL"]  
normal="these are multiple language used for coding and creating multiple apps and sits:-" 
funargs(normal,*text) 
```

```python
def argsfun(normal,*args,**kwargs):
  print(normal)
  for item in args:
    print(item)
  for key,value in kwargs.items():
     print(key,value) 
text=["python","Java","c","C++","ML","DL"]    
normal="these are multiple language used for coding and creating multiple apps and sits:-" 
thisdict = {"python": "high-level programming","name": "Guido van Rossum","year": 1991,"developed":"Python Software Foundation"
}
argsfun(normal,*text,**thisdict) 
```
