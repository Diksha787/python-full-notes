**# try and except**

```python
#The try block lets you test a block of code for errors.

#The except block lets you handle the error.

#The else block lets you execute code when thiere s no error.

#The finally block lets you execute code, regardless of the result of the try- and except blocks.
```

```python
for val in [12,3,4,0,6,7,8]:
  if 20/val > 0.3:
    print("yes")
  else:
    print("false")
```

```python
for val in [12,3,4,0,6,7,8]:
  try:
    if 20/val > 0.3:
      print("yes")
    else:
      print("false")

  except:
    print(" Error at {}".format(val))
```

**Lambda Function or Anonymous function
:-
Syntax(lambda arguments : expression)**

```python
def sum(a,b):
  return a+b
print(sum(3,4))  
```

```python
sum=lambda x,y:x+y
print(sum(3,4))
```

```python
a =[[3,5],[9,15],[2,46]]
a.sort()
```

```python
def f_name(a):
  return a[1] ## return values through the key index
a.sort(key=f_name)   ## key used as a function argument as a input
print(a)  
```

```python
a=[[1,14],[5,6],[3,18]]
a.sort(key=lambda x : x[1])  ## key =name of function (lambda)
print(a)
```

**Map:-The map() function executes a specified function for each item in an iterable. The item is sent to the function as a parameter.**

**Syntax:-map(function, iterables)**

```python
map(func , [12,55,66])
```

```python
for item in zip([1,2,3],[4,5,6]):
  print(item)
```

```python
numbers = ["13","31","42","20","15"]
for i in range(len(numbers)):
  numbers[i] = int(numbers[i])
numbers[2] = numbers[2]+numbers[1]  
print(numbers[2]) 
```

```python
def square(a):
  return a*a

def cube(a):
  return a*a*a 
  
func = [square, cube]
for i in range(5):
  val= list(map(lambda x : x(i),func))
  print(val)  
```

**Filter:-The filter() function returns an iterator were the items are filtered through a function to test if the item is accepted or not.**

**filter(function, iterable)**

```python
list_1=[1,2,3,4,5,6,7,8]

def is_greater_5(num):
  return num > 5
gre_than_5=list(filter(is_greater_5,list_1))
print(gre_than_5)
```
