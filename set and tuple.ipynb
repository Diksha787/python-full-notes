**Set Method**

**add(),clear(),copy(),difference(),difference_update(),discard(),intersection(),intersection_update(),pop(),remove(),symmetric_difference(),
symmetric_difference_update()	,union(),update()**

```python
## add Method:-Adds an element to the set
set = {12,25,12.5,"apple", "banana", "cherry",52,12}
set.add("Lichi")
print(set)
```

```python
## clear():-	Removes all the elements from the set
set = {12,25,12.5,"apple", "banana", "cherry",52,12}
set.clear()
set
```

```python
## difference():-	Returns a set containing the difference between two or more sets
set1 = {12,25,12.5,"apple", "banana", "cherry",52,12}
set2 = {15,32,25,"cherry",12,85,"apple"}

set3 = set2.difference(set1)
set3
```

```python
## difference_update():-	Removes the items in this set that are also included in another, specified set
set1 = {12,25,12.5,"apple", "banana", "cherry",52,12}
set2 = {15,32,25,"cherry",12,85,"apple"}
set2.difference_update(set1)
print(set1)
```

```python
## discard():-	Remove the specified item
set = {12,25,12.5,"apple", "banana", "cherry",52,12}
set.discard("banana")
print(set)
```

```python
## intersection():-	Returns a set, that is the intersection of two or more sets (return common element)
set1 = {12,25,12.5,"apple", "banana", "cherry",52,12}
set2 = {15,32,25,"cherry",12,85,"apple"}
set3 = set1.intersection(set2)
print(set3)
```

```python
##intersection_update():-	Removes the items in this set that are not present in other, specified set(s)
set1 = {12,25,12.5,"apple", "banana", "cherry",52,12}
set2 = {15,32,25,"cherry",12,85,"apple"}
set1.intersection_update(set2)
print(set1)
```

```python
##pop():-	Removes an element from the set
set = {12,25,12.5,"apple", "banana", "cherry",52,12}
set.pop()
```

```python
set = {12,25,12.5,"apple", "banana", "cherry",52,12}
set.pop(12)
```

```python
## remove():-	Removes the specified element
set = {12,25,12.5,"apple", "banana", "cherry",52,12}
set.remove(25)
print(set)
```

```python
## symmetric_difference()	Returns a set with the symmetric differences of two sets
set1 = {12,25,12.5,"apple", "banana", "cherry",52,12}
set2 = {15,32,25,"cherry",12,85,"apple"}
set1.symmetric_difference(set2)
```

```python
## symmetric_difference_update():-	inserts the symmetric differences from this set and another  (non duplicate values)
set1 = {12,25,12.5,"apple", "banana", "cherry",52,12}
set2 = {15,32,25,"cherry",12,85,"apple"}
set1.symmetric_difference_update(set2)
print(set1)
```

```python
## union():-	Return a set containing the union of sets
set1 = {12,25,12.5,"apple", "banana", "cherry",52,12}
set2 = {15,32,25,"cherry",12,85,"apple"}
set3 = set1.union(set2)
print(set3)
```

```python
## update()	Update the set with another set, or any other iterable
set1 = {12,25,12.5,"apple", "banana", "cherry",52,12}
set2 = {15,32,25,"cherry",12,85,"apple"}
set1.update(set2)
print(set1)
```

**Dictionary:- Dictionaries are written with curly brackets, and have keys and values:Dictionaries are used to store data values in key:value pairs.
A dictionary is a collection which is ordered*, changeable and do not allow duplicates.**

```python
thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964,
  "year": 2020
}
print(thisdict)
```

```python
print(len(thisdict))
```

```python
## change value
thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
thisdict["year"] = 2018
```

```python
## update value
thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
thisdict.update({"year": 2020})
print(thisdict)
```

```python
##clear()	Removes all the elements from the dictionary

car = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}

car.clear()

print(car)
```

```python
##copy()	Returns a copy of the dictionary
car = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}

x = car.copy()

print(x)
```

```python
##fromkeys()	Returns a dictionary with the specified keys and value
x = ('key1', 'key2', 'key3')

thisdict = dict.fromkeys(x)

print(thisdict)
```

```python
##get()	Returns the value of the specified key
car = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}

x = car.get("model")

print(x)
```

```python
##items()	Returns a list containing a tuple for each key value pair
car = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}

x = car.items()

print(x)
```

```python
##keys()	Returns a list containing the dictionary's keys
car = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}

x = car.keys()

print(x)
```

```python
##pop()	Removes the element with the specified key
car = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
x=car.pop("model")
print(x)
```

```python
##popitem()	Removes the last inserted key-value pair
car = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}

car.popitem()

print(car)
```

```python
##setdefault()	Returns the value of the specified key. If the key does not exist: insert the key, with the specified value
car = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}

x = car.setdefault("model", "Bronco")

print(x)
```

```python
##update()	Updates the dictionary with the specified key-value pairscar = {
car = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}

car.update({"color": "White"})

print(car)
```

```python
##values()	Returns a list of all the values in the dictionary
car = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}

x = car.values()

print(x)
```
