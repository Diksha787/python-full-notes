**2)Write a Python program to check if a set is a subset of another set.**

```python
setx = set(["apple", "mango"])
sety = set(["mango", "orange"])
setz = set(["mango"])

print("setx: ", setx)
print("sety: ", sety)
print("setz: ", setz)

print("Is setx a subset of sety: ", setx.intersection(sety) == setx)
print("Is sety a subset of setx: ", setx.intersection(sety) == sety)
print("Is setz a subset of setx: ", setx.intersection(setz) == setz)
```
