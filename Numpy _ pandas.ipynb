```python
# list of integers, convert to numpy array, check its dtype, type
import numpy as np
list_1=[12,45,67,45,43,8,53]

a=np.array(list_1)
print(a)
a.dtype
```

```python
a.dtype
```

```python
list_1=[12.7,45.98,67.76,45.67,43.34,8,53.12]
b=np.array(list_1)
print(b)
b.dtype
```

```python
list_1=["python","datascience","ML","Dl"]
c=np.array(list_1)
print(c)
c.dtype
```

```python
list_str = ["python","datascience","ML","Dl"]
arr_str = np.array(list_str)
print(arr_str.dtype)
print(type(arr_str))
```

```python
list_4 = [15,56.52,98,(34,17),('a',(5,2)),["hi",['hello',"greetings"]],{"a":"apple"}]
d = np.array(list_4)
print(d.dtype)
print(type(d))
```
